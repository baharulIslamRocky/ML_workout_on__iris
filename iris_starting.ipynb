```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris1=sns.load_dataset('iris')
```

```python
correlation= iris1.corr()
correlation
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(data=iris1.corr(),annot=True)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(data=iris1.corr(),cmap='coolwarm',annot=True)#cmap='YlGnBu'
```

```python
help(sns.heatmap)
```

```python
sns.lmplot(x="sepal_length", y="petal_length", hue="species", data = iris1, palette="Set1",\
           markers=["o", "s", "D"])
```

```python
#plt.figure(figsize=(10,8))

#sns.set(style="whitegrid", color_codes=True)
sns.lmplot(x='sepal_length',y='petal_length',data=iris1,hue="species", markers=["o", "s", "D"],palette="Set1")
```

```python
#plt.figure(figsize=(10,8))

#sns.set(style="whitegrid", color_codes=True)
sns.lmplot(x='sepal_length',y='petal_length',data=iris1,hue="species", markers=["o", "s", "D"])
```

```python
sns.lmplot(x='sepal_length',y='petal_length',data=iris1,hue='species',palette='Set1',markers=['o','s','D'],fit_reg=False)
```

```python
help(sns.lmplot)
```

```python
help(sns.pairplot)
```

```python
sns.pairplot(iris1,vars=['sepal_length','petal_length'],hue='species',markers=['o','s','D'],size=5)
```

```python
sns.pairplot(iris1,hue='species',markers=['o','s','D'])
```

```python
sns.set(style='whitegrid')
sns.pairplot(iris1,hue='species',markers=['o','s','D'],kind='reg',size=5)
```
