```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
iris=load_iris()
```

```python
x=iris.data
y=iris.target
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=0)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.4,random_state=0)
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(train_x,train_y)
accuracy_score(test_y,knn.predict(test_x))
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
accuracy_score(y_test,knn.predict(x_test))
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(train_x,train_y)
accuracy_score(test_y,knn.predict(test_x))
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
accuracy_score(y_test,knn.predict(x_test))
```

```python
def snsLine(x,y):
    plt.figure(figsize=(10,7))
    sns.lineplot(x=x,y=y)
    plt.xlabel('neighbors')
    plt.ylabel('Accuracy')
    plt.title('The accuracy line chart of Knn based on different neighbors')
```

```python
neighbors=list(range(1,26))
score=[]
for n in neighbors:
    knn=KNeighborsClassifier(n_neighbors=n)
    knn.fit(train_x,train_y)
    score.append(accuracy_score(test_y,knn.predict(test_x)))
   # print(n)
lineChart(neighbors,score)
snsLine(neighbors,score)
```
